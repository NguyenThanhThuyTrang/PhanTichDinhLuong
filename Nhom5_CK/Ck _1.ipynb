```python
country_mapping = {
    'Canada': 0,
    'United States': 1,
    'United Kingdom': 2,
    'France': 3,
    'Germany': 4,
    'Australia': 5
}

# Replace the country names with numbers
df['Region'] = df['Region'].map(country_mapping)

# Print the updated DataFrame
print(df)
```

```python
# Láº¥y táº­p há»£p cÃ¡c giÃ¡ trá»‹ duy nháº¥t tá»« cá»™t 'class'
unique_class = df['Class'].unique()

# Táº¡o má»™t tá»« Ä‘iá»ƒn Ã¡nh xáº¡ giá»¯a nÄƒm vÃ  mÃ£ sá»‘ nguyÃªn
class_to_integer = {Class: index for index, Class in enumerate(unique_class)}

# Sá»­ dá»¥ng map Ä‘á»ƒ mÃ£ hÃ³a cá»™t 'class'
df['class_encoded'] = df['Class'].map(class_to_integer)

# In ra káº¿t quáº£
print(df[['Class', 'class_encoded']])
```

```python
df
```

```python
# TrÃ­ch xuáº¥t nÄƒm tá»« cá»™t 'OrderDate'
df['Year'] = pd.to_datetime(df['OrderDate']).dt.year

# In DataFrame vá»›i cá»™t 'Year' má»›i
print(df['Year'])
```

```python
# Táº¡o má»™t báº£ng Ã¡nh xáº¡ tá»« nÄƒm sang giÃ¡ trá»‹ tÆ°Æ¡ng á»©ng
year_mapping = {2011: 0, 2012: 1, 2013: 2, 2014: 3}

# Thay tháº¿ cÃ¡c giÃ¡ trá»‹ nÄƒm báº±ng giÃ¡ trá»‹ tÆ°Æ¡ng á»©ng
df['Year'] = df['Year'].replace(year_mapping)

# In DataFrame vá»›i cá»™t 'Year' Ä‘Ã£ Ä‘Æ°á»£c thay tháº¿
print(df['Year'])
```
