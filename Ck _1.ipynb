```python
# Import required libraries
import pandas as pd
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
```

```python
df
```

```python
country_mapping = {
    'Canada': 0,
    'United States': 1,
    'United Kingdom': 2,
    'France': 3,
    'Germany': 4,
    'Australia': 5
}

# Replace the country names with numbers
df['Region'] = df['Region'].map(country_mapping)

# Print the updated DataFrame
print(df)
```

```python
# Lấy tập hợp các giá trị duy nhất từ cột 'class'
unique_class = df['Class'].unique()

# Tạo một từ điển ánh xạ giữa năm và mã số nguyên
class_to_integer = {Class: index for index, Class in enumerate(unique_class)}

# Sử dụng map để mã hóa cột 'class'
df['class_encoded'] = df['Class'].map(class_to_integer)

# In ra kết quả
print(df[['Class', 'class_encoded']])
```

```python
df
```

```python
# Trích xuất năm từ cột 'OrderDate'
df['Year'] = pd.to_datetime(df['OrderDate']).dt.year

# In DataFrame với cột 'Year' mới
print(df['Year'])
```

```python
# Tạo một bảng ánh xạ từ năm sang giá trị tương ứng
year_mapping = {2011: 0, 2012: 1, 2013: 2, 2014: 3}

# Thay thế các giá trị năm bằng giá trị tương ứng
df['Year'] = df['Year'].replace(year_mapping)

# In DataFrame với cột 'Year' đã được thay thế
print(df['Year'])
```

```python
df
```

```python
df.drop(['ProductID','ProductID','CustomerID','SalesOrderID','OrderDate', 'StoreName','Class'],axis=1,inplace=True)
```

```python
# Số hóa các biến liên tục: UnitPrice, Totalsale
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['UnitPrice'] = scaler.fit_transform(df['UnitPrice'].values.reshape(-1, 1))
df['Totalsale'] = scaler.fit_transform(df['Totalsale'].values.reshape(-1, 1))
```
